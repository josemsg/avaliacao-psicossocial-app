```python
import streamlit as st
import pandas as pd
from datetime import datetime
import os

st.set_page_config(page_title="Avalia√ß√£o Psicossocial - PGR", layout="centered")

st.title("üìã Avalia√ß√£o de Riscos Psicossociais - PGR")
st.markdown("Preencha o formul√°rio abaixo com sinceridade. Seus dados s√£o confidenciais e utilizados exclusivamente para fins de avalia√ß√£o no Programa de Gerenciamento de Riscos.")

# Entradas do formul√°rio
nome = st.text_input("Nome Completo")
cargo = st.text_input("Cargo/Fun√ß√£o")
setor = st.text_input("Setor")
tempo_funcao = st.number_input("Tempo na fun√ß√£o (em anos)", min_value=0)

st.subheader("Avalia√ß√£o de Riscos Psicossociais")
pressao = st.selectbox("Voc√™ sente press√£o no trabalho?", ["Nunca", "Raramente", "Frequentemente", "Sempre"])
carga = st.selectbox("Voc√™ tem carga de trabalho excessiva?", ["N√£o", "√Äs vezes", "Frequentemente", "Sempre"])
assedio = st.selectbox("J√° sofreu ass√©dio moral ou sexual?", ["Nunca", "Raramente", "Frequentemente", "Sempre"])

# Convers√£o das respostas para pontua√ß√£o
mapa = {"Nunca": 1, "N√£o": 1, "Raramente": 2, "√Äs vezes": 2, "Frequentemente": 3, "Sempre": 4}
risco_total = mapa[pressao] + mapa[carga] + mapa[assedio]

# Classifica√ß√£o do n√≠vel de risco
if risco_total <= 4:
    nivel_risco = "Toler√°vel"
elif 5 <= risco_total <= 7:
    nivel_risco = "Moderado"
elif 8 <= risco_total <= 10:
    nivel_risco = "Substancial"
else:
    nivel_risco = "Intoler√°vel"

st.success(f"N√≠vel de risco identificado: **{nivel_risco}**")

# Bot√£o para envio dos dados
if st.button("Enviar Resposta"):
    resposta = {
        "Data": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        "Nome": nome,
        "Cargo": cargo,
        "Setor": setor,
        "Tempo na fun√ß√£o (anos)": tempo_funcao,
        "Press√£o no trabalho": pressao,
        "Carga excessiva": carga,
        "Ass√©dio": assedio,
        "N√≠vel de risco": nivel_risco
    }

    file_path = "responses.csv"
    file_exists = os.path.isfile(file_path)

    df = pd.DataFrame([resposta])
    df.to_csv(file_path, mode='a', header=not file_exists, index=False)

    st.success("‚úÖ Resposta enviada com sucesso! Obrigado por participar da avalia√ß√£o.")
```
